```python
# Importing libraries needed
import numpy as np  # for matrices, array, linear algebra
import pandas as pd  # open source data analysis and manipulation
from numpy import *
import scipy
import scipy.linalg 
import matplotlib.pyplot as plt
```

```python
# Importing the Dataset
regression=pd.read_csv("hw8q4.csv")

print(regression)
```

```python
x = regression.x.values
y = regression.y.values

# Training set (odd indices) and testing set (even indices)
x_train = x[::2]
y_train = y[::2]
x_test = x[1::2]
y_test = y[1::2]
```

```python
x_linear_train = np.vstack([np.ones(x_train.shape[0]), x_train]).T
theta_linear = np.linalg.solve(x_linear_train.T @ x_linear_train, x_linear_train.T @ y_train)
print("Linear solution: theta_0 =", theta_linear[0], ", theta_1 =", theta_linear[1])
```

```python
linear_rss_train = np.linalg.norm(x_linear_train @ theta_linear - y_train) ** 2 / len(y_train)
print("Training Linear Mean RSS =", linear_rss_train)

x_linear_test = np.vstack([np.ones(x_test.shape[0]), x_test]).T
y_test_pred = x_linear_test @ theta_linear
linear_rss_test = np.linalg.norm(y_test_pred - y_test) ** 2 / len(y_test)
print("Testing Linear Mean RSS =", linear_rss_test)
```

```python
x_values = np.linspace(min(x), max(x), 100)
x_linear_values = np.vstack([np.ones(x_values.shape[0]), x_values]).T
y_linear_values = x_linear_values @ theta_linear

plt.scatter(x_train, y_train, color="blue", label="Training Data")
plt.scatter(x_test, y_test, color="red", label="Testing Data")
plt.plot(x_values, y_linear_values, color="green", label="Linear Model")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Model with Training and Testing Data")
plt.legend()
plt.show()
```

```python
x_quad_train = np.vstack([np.ones(x_train.shape[0]), x_train, x_train**2]).T
theta_quad = np.linalg.solve(x_quad_train.T @ x_quad_train, x_quad_train.T @ y_train)
print("Quadratic solution: theta_0 =", theta_quad[0], ", theta_1 =", theta_quad[1], ", theta_2 =", theta_quad[2])
```

```python
quad_rss_train = np.linalg.norm(x_quad_train @ theta_quad - y_train) ** 2 / len(y_train)
print(f"Training Quadratic Mean RSS: {quad_rss_train}")

x_quad_test = np.vstack([np.ones(x_test.shape[0]), x_test, x_test**2]).T
y_test_pred = x_quad_test @ theta_quad
quad_rss_test = np.linalg.norm(y_test_pred - y_test) ** 2 / len(y_test)
print(f"Testing Quadratic Mean RSS: {quad_rss_test}")
```

```python
x_quad_values = np.vstack([np.ones(x_values.shape[0]), x_values, x_values**2]).T
y_quad_values = x_quad_values @ theta_quad

plt.scatter(x_train, y_train, color="blue", label="Training Data")
plt.scatter(x_test, y_test, color="red", label="Testing Data")
plt.plot(x_values, y_quad_values, color="green", label="Quadratic Model")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Regression Model with Training and Testing Data")
plt.legend()
```

```python
x_quint_train = np.vstack([np.ones(x_train.shape[0]), x_train, x_train**2, x_train**3, x_train**4, x_train**5]).T
theta_quint = np.linalg.solve(x_quint_train.T @ x_quint_train, x_quint_train.T @ y_train)
print("Quintuple solution: theta_0 =", theta_quint[0], ", theta_1 =", theta_quint[1], ", theta_2 =", theta_quint[2], 
      ", theta_3 =", theta_quint[3], ", theta_4 =", theta_quint[4], ", theta_5 =", theta_quint[5])
```

```python
quint_rss_train = np.linalg.norm(x_quint_train @ theta_quint - y_train) ** 2 / len(y_train)
print(f"Training Quintuple Mean RSS: {quint_rss_train}")

x_quint_test = np.vstack([np.ones(x_test.shape[0]), x_test, x_test**2, x_test**3, x_test**4, x_test**5]).T
y_test_pred = x_quint_test @ theta_quint
quint_rss_test = np.linalg.norm(y_test_pred - y_test) ** 2 / len(y_test)
print(f"Testing Quintuple Mean RSS: {quint_rss_test}")
```

```python
x_quint_values = np.vstack([np.ones(x_values.shape[0]), x_values, x_values**2, x_values**3, x_values**4, x_values**5]).T
y_quint_values = x_quint_values @ theta_quint

plt.scatter(x_train, y_train, color="blue", label="Training Data")
plt.scatter(x_test, y_test, color="red", label="Testing Data")
plt.plot(x_values, y_quint_values, color="green", label="Quadratic Model")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Regression Model with Training and Testing Data")
plt.legend()
```

```python
folds = [
    {
        "train_x": np.array([3, 5, 9, 11]),
        "train_y": np.array([7.5, 9.5, 8, 3.5]),
        "test_x": np.array([1, 7]),
        "test_y": np.array([4, 10])
    },
    {
        "train_x": np.array([1, 5, 7, 11]),
        "train_y": np.array([4, 9.5, 10, 3.5]),
        "test_x": np.array([3, 9]),
        "test_y": np.array([7.5, 8])
    },
    {
        "train_x": np.array([1, 3, 7, 9]),
        "train_y": np.array([4, 7.5, 10, 8]),
        "test_x": np.array([5, 11]),
        "test_y": np.array([9.5, 3.5])
    }
]
```

```python
fold1_train_x = folds[0]['train_x']
fold1_train_y = folds[0]['train_y']
fold1_test_x = folds[0]['test_x']
fold1_test_y = folds[0]['test_y']
X_train = np.vstack([np.ones(fold1_train_x.shape[0]), fold1_train_x]).T
X_test = np.vstack([np.ones(fold1_test_x.shape[0]), fold1_test_x]).T
theta_linear = np.linalg.solve(X_train.T @ X_train, X_train.T @ fold1_train_y)
y_train_pred = X_train @ theta_linear
y_test_pred = X_test @ theta_linear
mse_train = np.mean((y_train_pred - fold1_train_y) ** 2)
mse_test = np.mean((y_test_pred - fold1_test_y) ** 2)
print(f"Fold 1 - Linear Model:")
print(f"  Training MSE: {mse_train}")
print(f"  Testing MSE: {mse_test}")

plt.scatter(fold1_train_x, fold1_train_y, color='blue', label='Train Data')
plt.scatter(fold1_test_x, fold1_test_y, color='red', label='Test Data')
plt.plot(fold1_train_x, y_train_pred, color='green', label='Linear Fit (Train)')
plt.title('Linear Model for Fold 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
mse_linear = []
mse_quadratic = []
mse_cubic = []

for i in range(len(x_train)):
    x_train_loo = np.delete(x_train, i)
    y_train_loo = np.delete(y_train, i)
    x_test_loo = x_train[i]
    y_test_loo = y_train[i]
    
    # Linear model
    X_train_linear = np.vstack([np.ones(x_train_loo.shape[0]), x_train_loo]).T
    theta_linear = np.linalg.solve(X_train_linear.T @ X_train_linear, X_train_linear.T @ y_train_loo)
    y_pred_linear = np.array([1, x_test_loo]) @ theta_linear
    mse_linear.append((y_pred_linear - y_test_loo) ** 2)
    
    # Quadratic model
    X_train_quadratic = np.vstack([np.ones(x_train_loo.shape[0]), x_train_loo, x_train_loo**2]).T
    theta_quadratic = np.linalg.solve(X_train_quadratic.T @ X_train_quadratic, X_train_quadratic.T @ y_train_loo)
    y_pred_quadratic = np.array([1, x_test_loo, x_test_loo**2]) @ theta_quadratic
    mse_quadratic.append((y_pred_quadratic - y_test_loo) ** 2)
    
    # Cubic model
    X_train_cubic = np.vstack([np.ones(x_train_loo.shape[0]), x_train_loo, x_train_loo**2, x_train_loo**3]).T
    theta_cubic = np.linalg.solve(X_train_cubic.T @ X_train_cubic, X_train_cubic.T @ y_train_loo)
    y_pred_cubic = np.array([1, x_test_loo, x_test_loo**2, x_test_loo**3]) @ theta_cubic
    mse_cubic.append((y_pred_cubic - y_test_loo) ** 2)

avg_mse_linear = np.mean(mse_linear)
avg_mse_quadratic = np.mean(mse_quadratic)
avg_mse_cubic = np.mean(mse_cubic)

print(f"Leave-One-Out CV - Linear Model Average MSE: {avg_mse_linear}")
print(f"Leave-One-Out CV - Quadratic Model Average MSE: {avg_mse_quadratic}")
print(f"Leave-One-Out CV - Cubic Model Average MSE: {avg_mse_cubic}")
```
